```python
con = sqlite3.connect("my_data1.db")
cur = con.cursor()
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

### Task 2

```python
%sql select * from SPACEXTBL where Launch_Site like '%CCA%' limit 5
```

### Task 3

```python
%sql select  Customer, SUM(PAYLOAD_MASS__KG_) as "total payload mass" from SPACEXTBL where Customer = 'NASA (CRS)' 
```

### Task 4

```python
%sql select Booster_Version, AVG(PAYLOAD_MASS__KG_) as "Average Playload Mass" from SPACEXTBL where Booster_Version like '%F9 v1.1%'
```

### Task 5

```python
%sql select MIN(Date) as Date from SPACEXTBL where Landing_Outcome = 'Success (ground pad)'
```

### Task 6

```python
%sql select Booster_Version, Landing_Outcome, PAYLOAD_MASS__KG_ from SPACEXTBL where PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000 and Landing_Outcome = 'Success (drone ship)'
```

### Task 7

```python
%sql select COUNT(Mission_Outcome), Mission_Outcome from SPACEXTBL GROUP BY Mission_Outcome
```

### Task 8

```python
%sql select Booster_Version, MAX(PAYLOAD_MASS__KG_) from SPACEXTBL GROUP BY Booster_Version 
```

### Task 9

```python
%sql select SUBSTR(Date, 6,2) as month, Booster_Version, Launch_Site, Landing_Outcome from SPACEXTBL where SUBSTR(Date,0,5)='2015' and Landing_Outcome = 'Failure (drone ship)'
```

### Task 10
